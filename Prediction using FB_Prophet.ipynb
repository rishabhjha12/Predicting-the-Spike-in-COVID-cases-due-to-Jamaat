```
#Learn how to read a .xls file by creating a dataframe using pandas
# Reading the datasets
df= pd.read_excel('lock.xlsx')
df_india = df.copy()
df
```

```
from fbprophet import Prophet
```

```
confirmed = df.groupby('Date').sum()['Total Cases'].reset_index()
```

```

confirmed
```

```
confirmed.columns = ['ds','y']
```

```
m = Prophet(interval_width=0.95)
m.fit(confirmed)
future = m.make_future_dataframe(periods=45)
future
```

```
#predicting the future with date, and upper and lower limit of y value
forecast = m.predict(future)
forecast[['ds',  'yhat_upper']]
```
